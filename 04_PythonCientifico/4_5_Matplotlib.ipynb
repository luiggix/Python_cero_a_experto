### Dibujar la función senoidal para 100 puntos en el intervalo de 0 a 2pi.

El comando básico, para graficar funciones de una variable (y = f(x)), es **plot(x,y)**, donde x e y, son arreglos o listas de tipo númerico del mismo tamaño.

```python
import numpy as np # Importamos el modúlo numpy
import matplotlib.pyplot as plt # Importamos el modulo pyplot de numpy
# Generamos un arreglo de 100 
# puntos que va desde 0 a 2pi
x = np.linspace(0,2*np.pi,100) # Dominio
# Evaluamos cada punto del dominio
# en la funcion senoidal y lo almacenamos
y = np.sin(x) # Rango
# Graficamos el resultado
plt.plot(x,y)
# Mostramos en pantalla
plt.show()
```

###  Dibujamos la función senoidal y cosenoidal en la misma ventana.

```python
import numpy as np # Importamos el modúlo numpy
import matplotlib.pyplot as plt # Importamos el modulo pyplot de numpy
# Generamos un arreglo de 100 
# puntos que va desde 0 a 2pi
x = np.linspace(0,2*np.pi,100) # Dominio
# Evaluamos cada punto del dominio
# en la funcion senoidal y lo almacenamos
y = np.sin(x) # Rango
# Evaluamos cada punto del dominio
# en la funcion cosenoidal y lo almacenamos
f = np.cos(x) # Rango 2
# Graficamos el resultado
plt.plot(x,y) # Graficamos la función senoidal
plt.plot(x,f) # Graficamos la función cosenoidal
# Mostramos en pantalla
plt.show()
```

###  Dibujamos en pantalla la función cosenoidal y senoidal en distintas figuras.

```python
import numpy as np
import matplotlib.pyplot as plt

# Creamos dos figuras distintas en donde graficar
plt.figure(1)   # Identificador 1
plt.figure('dos') # Identificador dos
# Generamos el dominio de 0 a 2pi con 100 puntos
x = np.linspace(0,2*np.pi,100)
# Hacemos la evaluacion de estos puntos
y = np.sin(x)
f = np.cos(x)
# Dibujamos la funcion coseno en la ventada dos
plt.figure('dos')
plt.plot(x,f)
# Dibujamos la funcion seno en grafica 1
plt.figure(1)
plt.plot(x,y)
# Mostramos en pantalla el resultado
plt.show()
```

### Dibuja la función senoidal y cosenodial en la misma ventana. 

También es posible dibujar los gráficos separados en la misma ventana, esto por medio del comando subplot(nrows, ncols, ngraf), donde nrows y ncols indica en cuantas lineas y columnas se divide la misma ventana, ngraf indica de que gráfica.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generamos el dominio de 0 a 2pi con 100 puntos
x = np.linspace(0,2*np.pi,100)
# Hacemos la evaluacion de estos puntos
y = np.sin(x)
f = np.cos(x)
# Dibujamos la funcion coseno en la primera
# mitad de la grafica
plt.subplot(1,2,1)
plt.plot(x,y)
# Dibujamos la funcion seno en la segunda
# mitad de la grafica
plt.subplot(1,2,2)
plt.plot(x,f)
plt.savefig('grafica.pdf')
plt.show()
```

## Personalización


Matplotlib, permite modificar la apariencia de sus gráficos, con esto el usuario puede dar una mayor claridad a sus resultados.

Algunas de las características que se pueden modificar en los gráficos son:

- **Tipo de línea.**
- **Espesor.**
- **Marcador.**
- **Color.**


## Mostrar en pantalla la función senoidal y cosenoidal con leyendas y una malla. 

```python
import numpy as np
import matplotlib.pyplot as plt
# Generamos el dominio
x = np.linspace(0,2*np.pi,100)
# Mostramos en pantalla la grafica del seno
plt.plot(x,np.sin(x),label="sin(x)")
# Mostramos en pantalla la grafica de coseno
plt.plot(x,np.cos(x),label="cos(x)")
# Agregamos las leyendas
plt.legend()
# Agregamos una malla al grafico
plt.grid()
# Etiqueta a los ejes
plt.xlabel("x")
plt.ylabel("Funciones Trigonometricas")
plt.show()
```

## Gráficar la función senoidal de 0 a 2pi con 100 puntos y la linea discontinua.

**Tipo de línea**

En python los gráficos son valores discretos que se unen por defecto por medio de línes continuas.

```python
# Agregamos la libreria de numpy
import numpy as np
# Agregamos la librieria matplotlib
import matplotlib.pyplot as plt
# Generamos el dominio con linspace
x = np.linspace(0,2*np.pi,100)
# Graficamos la funcion
plt.plot(x,np.sin(x),linestyle='dashed')
# Mostramos en pantalla
plt.show()
```

## Gráficar la función senoidal de 0 a 2pi con 100 puntos y la linea con un espesor mayor.


**Espesor**

El espesor de la lineas puede modificarse fácilmente desde el comando plot, añadiendo el parametro linewidth.

```python
# Agregamos la libreria de numpy
import numpy as np
# Agregamos la librieria matplotlib
import matplotlib.pyplot as plt
# Generamos el dominio con linspace
x = np.linspace(0,2*np.pi,100)
# Graficamos la funcion
plt.plot(x,np.sin(x),linewidth=6)
# Mostramos en pantalla
plt.show()
```

## Gráficar la función senoidal de 0 a 2pi con 15 puntos y un marcador triangular.

**Marcador**

Los marcadores son la forma de los puntos que se están mostrando en pantalla.

```python
# Agregamos la libreria de numpy
import numpy as np
# Agregamos la librieria matplotlib
import matplotlib.pyplot as plt
# Generamos el dominio con linspace
x = np.linspace(0,2*np.pi,15)
# Graficamos la funcion
plt.plot(x,np.sin(x),marker='v')
# Mostramos en pantalla
plt.show()
```

## Graficar la función senoidal y cosenoidal usando 20 puntos de 0 a 2pi y añadiendo todos los parametros vistos hasta ahora. 

```python
# Agregamos la libreria de numpy
import numpy as np
# Agregamos la librieria matplotlib
import matplotlib.pyplot as plt
# Generamos el dominio con linspace
x = np.linspace(0,2*np.pi,20)
# Graficamos la funcion seno
plt.plot(x,np.sin(x),'r:',label='$sin(x)$')
# Graficamos la funcion coseno
plt.plot(x,np.cos(x),marker='p',linewidth=3,label='$cos(x)$')
# Etiqueta del eje x
plt.xlabel('x')
# Etiqueta del eje y
plt.ylabel('Funciones trigonometricas')
# Mostramos la leyenda
plt.legend()
# Mostramos la malla
plt.grid()
# Mostramos en pantalla
plt.show()
```

## Contornos


Hasta ahora trabajamos con funciones reales de variable real (y = f(x)). Sin embargo, hay varias situaciones en las cuales se requiere visualizar funciones reales de variable vectorial(z = f (x, y)). Una manera de visualizar una función de dos variables consiste en utilizar sus curvas de nivel o líneas de contorno. Las curvas de nivel ayudan a visualizar funciones tridimensionales en un espacio bidimensional.

##  Obtener las curvas de nivel para la gráfica f(x,y) = cos(x)sin(x)


```python
# Agregamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
# Generamos el dominio
x = np.linspace(-5,5,20)
y = np.linspace(-5,5,20)
# Generamos la malla
X,Y = np.meshgrid(x,y)
# Graficamos la curva
plt.contour(X,Y,np.cos(X)*np.sin(Y))
# Mostramos el resultado
plt.show()
```

## Ejemplo 2.  

```python
# Agregamos las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
# Generamos el dominio
x = np.linspace(-5,5,20)
y = np.linspace(-5,5,20)
# Generamos la malla
X,Y = np.meshgrid(x,y)
# Graficamos la curva
plt.contourf(X,Y,np.cos(X)*np.sin(Y))
# Mostramos el resultado
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.viridis)

plt.show()
```
